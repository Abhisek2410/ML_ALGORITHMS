```python
import pandas as pd
path="C:/Users/abhis/Desktop/ML/salaries.csv"
data=pd.read_csv(path)
print(data)
print(data.info())

```

```python
 import sklearn
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()

data['company_new']=le_company.fit_transform(data['company'])
data['job_new']=le_company.fit_transform(data['job'])
data['degree_new']=le_company.fit_transform(data['degree'])
print(data)

inputs=data.drop(['company','job','degree','salary_more_then_100k'],'columns')
output=data.drop(['company','company_new','job','degree','job_new','degree_new'],'columns')
print(inputs)
print(output)

from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(inputs,output)
result=model.predict([[2,0,0]])
print(result)
```
